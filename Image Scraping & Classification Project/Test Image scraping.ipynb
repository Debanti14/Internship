```python
#Importing required libraries

import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

import selenium
from selenium import webdriver
import re
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

import time
from bs4 import BeautifulSoup

from selenium.common.exceptions import ElementNotInteractableException
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException

import requests
```

```python
#Connect to web driver

driver = webdriver.Chrome(r"C:\Users\HP\Downloads\chromedriver_win32 (2)\chromedriver.exe")
```

```python
# Opening the Amazon.in
driver.get('https://www.flipkart.com/search?q=sarees&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off')
```

```python
Saree_name = []
sareeimg_url=[]


for i in range(0,10):
    for j in driver.find_elements(By.XPATH, "//a[@class='IRpwTa']"):
        Saree_name.append(j.text)
    for j in driver.find_elements(By.XPATH, "//img[@class='_2r_T1I']"):
        sareeimg_url.append(j.get_attribute('src'))
        
    #Path for next page as it changes for every page. We are appending numbers as pages change 
    k=i+1
    next_page="https://www.flipkart.com/search?q=sarees&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"+str(k)
    driver.get(next_page)
```

```python
# Checking the length of data extracted for sarees
print(len(sareeimg_url))
```

```python
for i in range(len(sareeimg_url)):

    if i >= 250:

        break

    print("Downloading {0} of {1} images" .format(i, 250))

    response= requests.get(sareeimg_url[i])

    file = open(r"D:\Images\Test\Sarees\saree"+str(i)+".jpg", "wb")

    file.write(response.content)
```

```python
# Getting the website to driver
driver.get('https://www.amazon.in/s?k=amazon+jeans+for+men&page=6&hvadid=72224283643800&hvbmt=be&hvdev=c&hvqmt=e&qid=1668237326&ref=sr_pg_6')
```

```python
Jeans_name = []
jeansimg_url=[]


for i in range(0,5):
    for j in driver.find_elements(By.XPATH, "//span[@class='a-size-base-plus a-color-base a-text-normal']"):
        Jeans_name.append(j.text)
    for j in driver.find_elements(By.XPATH, "//img[@class='s-image']"):
        jeansimg_url.append(j.get_attribute('src'))
    
    #Path for next page as it changes for every page. We are appending numbers as pages change 

    k=i+1
    next_page="https://www.amazon.in/s?k=amazon+jeans+for+men&page=6&hvadid=72224283643800&hvbmt=be&hvdev=c&hvqmt=e&qid=1668237326&ref=sr_pg_6"+str(k)
    driver.get(next_page)
```

```python
# Checking the length of data extracted for jeans
print(len(jeansimg_url))
```

```python
for i in range(len(jeansimg_url)):

    if i >= 250:

        break

    print("Downloading {0} of {1} images" .format(i, 250))

    response= requests.get(jeansimg_url[i])

    file = open(r"D:\Images\Test\Jeans\jeans"+str(i)+".jpg", "wb")

    file.write(response.content)
```
