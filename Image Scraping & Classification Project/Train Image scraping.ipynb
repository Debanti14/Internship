```python
#Importing required libraries

import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

import selenium
from selenium import webdriver
import re
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

import time
from bs4 import BeautifulSoup

from selenium.common.exceptions import ElementNotInteractableException
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException

import requests
```

```python
#Connect to web driver

driver = webdriver.Chrome(r"C:\Users\HP\Downloads\chromedriver_win32 (2)\chromedriver.exe")
```

```python
# Opening the Amazon.in
driver.get('https://www.amazon.in/')
```

```python
# Searching laptops in the search bar and clicking the search button
search_bar=driver.find_element(By.ID,'twotabsearchtextbox')
search_bar.send_keys("sarees")
```

```python
driver.find_element(By.ID,'nav-search-submit-button').click()
```

```python
Saree_name = []
sareeimg_url=[]


for i in range(0,10):
    for j in driver.find_elements(By.XPATH, "//span[@class='a-size-base-plus a-color-base a-text-normal']"):
        Saree_name.append(j.text)
    for j in driver.find_elements(By.XPATH, "//img[@class='s-image']"):
        sareeimg_url.append(j.get_attribute('src'))
    
#Path for next page as it changes for every page. We are appending numbers as pages change 
k=i+1
next_page=driver.find_element(By.XPATH,"//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']"+str(k))
next_page.click()

```

```python
# Checking the length of data extracted for sarees
print(len(sareeimg_url))
```

```python
for i in range(len(sareeimg_url)):

    if i >= 250:

        break

    print("Downloading {0} of {1} images" .format(i, 250))

    response= requests.get(sareeimg_url[i])

    file = open(r"D:\Images\Train\Sarees\saree"+str(i)+".jpg", "wb")

    file.write(response.content)
```
