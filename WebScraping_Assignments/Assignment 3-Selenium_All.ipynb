```python
prod_opening_url=[]


for page in range(3):
    
    url=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    
    for i in url:
       
        prod_opening_url.append(i.get_attribute('href'))
    
    nxt_button=driver.find_elements(By.XPATH,'/html/body/div[1]/div[2]/div[1]/div[1]/div/span[3]/div[2]/div[57]/div/div/span/a[8]')     #scraping the list of buttons from the page
    
    time.sleep(3)
        
    nxt_button
```

```python
# blank list variables   

product = []
brand_name = []
prod_price = []
return_exchange = []
del_tag = []
availability = []


time.sleep(3)

for url in prod_opening_url:   ##loop for every guitar in the list

    driver.get(url)

    time.sleep(2)

    try:

        product.append(driver.find_elements(By.XPATH,'//span[@id="productTitle"]').text)
        brand_name.append(driver.find_elements(By.XPATH,'//span[@class="a-size-base"]').text)
        prod_price.append(driver.find_elements(By.XPATH,'//span[@class="a-price-whole"]').text)
        return_exchange.append(driver.find_elements(By.XPATH,'//a[@class="a-size-small a-link-normal a-text-normal"]').text)
        del_tag.append(driver.find_elements(By.XPATH,'//span[@class="a-text-bold"]').text)
        availability.append(driver.find_elements(By.XPATH,'//span[@class="a-size-medium a-color-success"]').text)

    except NoSuchElementException:
             
        product.append('-')
        brand_name.append('-')
        prod_price.append('-')
        return_exchange.append('-')
        del_tag.append('-')
        availability.append('-')
    
```
