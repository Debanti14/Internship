# 1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
driver = webdriver.Chrome()
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
# give some time to load the webpage
time.sleep(3)
```

```python
#scraping rank of the video

rank=[]

try:

    ranks=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]")

    for i in ranks:

        rank.append(i.text)

except NoSuchElementException:#handling no such element exception

    rank.append('No details available')

except StaleElementReferenceException:#handling Stale element exception

    rank.append('No details available')
```

```python
rank
```

```python
names=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]//tbody//tr//td[2]')

name=[]

for i in names[0:30]:

       name.append(i.text)
name   
```

```python
artist=[]

artists=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]//tbody//tr//td[3]')

for i in artists[0:30]:

       artist.append(i.text)
        
artist
```

```python
upload_dt=[]

upload_dates=driver.find_elements(By.XPATH,'//TABLE[@class="wikitable sortable jquery-tablesorter"]//tbody//tr//td[5]')

for i in upload_dates[0:30]:

       upload_dt.append(i.text)
        
upload_dt
```

```python
vw_tag=driver.find_elements(By.XPATH,'//TABLE[@class="wikitable sortable jquery-tablesorter"]//tbody//tr//td[4]')

views=[]

for i in vw_tag[0:30]:
    
    views.append(i.text)
    
views 
```

```python
video_data=pd.DataFrame()

video_data['Rank'] = rank
video_data['Video Name'] = name
video_data['Artist'] = artist
video_data['Uploaded on'] = upload_dt
video_data['Total Views (in Billions)'] = views
```

```python
video_data
```

```python
video_data.to_csv('Top 30 most-viewed YouTube videos')
```

```python
driver.close()
```
